```python
from geopy.geocoders import Nominatim

def create_app():
    geolocator = Nominatim(user_agent="geo_app")
    
    return geolocator

geo_app = create_app()
```

```python
import pandas as pd
import time

pd.options.mode.chained_assignment = None

def geo_search(endereco):
    location = geo_app.geocode(endereco['logradouro'] + ", " + endereco['cidade'] + " - " + endereco['bairro'])
        
    lat = location.latitude
    long = location.longitude
    
    print(location)
    
    return lat, long

def load_data():
    df = pd.read_csv("data/raw/zonas/lista_zonas_eleitorais.csv", encoding="Latin-1")
    
    return df

def execute(geo_app):
    # establishing address pattern
    endereco = {
        "logradouro":'',
        "cidade":'',
        "bairro":''
    }
    
    zdf = load_data()
    
    zdf["latitude"] = 0.00
    zdf["longitude"] = 0.00
    
    for i in zdf.index:
        #here we will overwrite the value of the dictionary
        endereco["logradouro"] = zdf["endereco"].loc[i]
        endereco["cidade"] = zdf["nome_municipio"].loc[i]
        endereco["bairro"] = zdf["bairro"].loc[i]
        
        try:
            time.sleep(5)
            lat, lon = geo_search(endereco)
            
            zdf["latitude"].loc[i] = lat
            zdf["longitude"].loc[i] = lon
        except:
            continue
    
    return zdf
```

```python
zdf = execute(geo_app)
```
